# **Import Libraries**

```python
import pandas as pd
import pickle
import numpy as np
from sklearn.metrics import classification_report
from IPython.display import display, HTML
```

# **Data Loading**

```python
'''Import data dari csv'''

df = pd.read_csv('inference_data.csv', delimiter=',')
df.head()
```

```python
'''Menghitung proporsi kelas minoritas dan mayoritas untuk menentukan penggunaan model'''

# Menghitung jumlah masing-masing nilai pada kolom 'y_encoded'
value_counts = df['y_encoded'].value_counts()
print("Jumlah masing-masing nilai pada kolom 'y_encoded':")
print(value_counts)

# Menghitung persentase kelas minoritas (nilai 1)
total_count = len(df)
minority_class_count = value_counts.get(1, 0)
minority_class_percentage = (minority_class_count / total_count) * 100

print(f"Persentase kelas minoritas (1) pada kolom 'y_encoded': {minority_class_percentage:.2f}%")
```

Berdasarkan perhitungan diketahui bahwa proporsi kelas minoritas (1) pada data hanya 11,72% dari total data yang berarti data ini terindikasi imbalance, sehingga model yang akan digunakan untuk data ini adalah model yang sudah melalui oversampling dan hyperparameter tuning.

```python
'''Memuat model dari file'''

with open('xgb_pipeline_tuning_smote.pkl', 'rb') as file_model:
    grid_search_smote_tuned_loaded = pickle.load(file_model)
```

```python
'''Prediksi'''
predictions = grid_search_smote_tuned_loaded.predict(df)
```

```python
'''Menambahkan hasil prediksi ke DataFrame'''
df['Predictions'] = predictions
```

```python
'''Membuat tampilan scrollable untuk DataFrame agar lebih mudah dibaca'''

html = df.to_html()
scrollable_html = f"""
<div style="height:500px; overflow:auto; border:1px solid black; padding:10px">
{html}
</div>
"""

# Menampilkan DataFrame yang telah digabungkan dengan hasil prediksi
display(HTML(scrollable_html))
```

```python
'''Menampilkan Classification Report'''

y_true = df['y_encoded']

# Menampilkan Classification Report
print("Classification Report:")
print(classification_report(y_true, predictions))
```
